```python
import pandas as pd
```

```python
df=pd.read_csv("NorthCarolina Data.csv").drop("Unnamed: 0",axis=1)
```

```python
df.columns
```

```python
df[df.contest_title=='US PRESIDENT']['election_dt']
```

```python
dat = pd.concat([df['county'],df['contest_title'],df['precinct_name'],df['precinct_code'],df['vote_ct']],axis=1)
```

```python
df_us=dat[dat.contest_title=='US PRESIDENT'].dropna(how="any")
df_us
```

```python
dat_us = df_us.groupby('precinct_name').agg({
    # preparing for turnout
    'vote_ct': 'sum',
    # Polling Station level anyway
    'contest_title': 'first',
    'precinct_code': 'first',
    'county': 'first'
})

```

```python
dat_us.drop('contest_title',axis=1,inplace=True)
```

```python
dat_us.index.name = 'precinct_name'
dat_us.reset_index(inplace=True)
# replace the / for a - to be in sync with the image names
dat_us['precinct_name'] = dat_us['precinct_name']
```
